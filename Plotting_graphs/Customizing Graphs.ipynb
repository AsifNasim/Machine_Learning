```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.1) # starting element, ending element and interval
y = x ** 3
# plt.plot(x,y, color ="black", marker = "^")
plt.plot(x,y, color ="black", linewidth = "15") # with width of the desired line

plt.show()
```

# Label, Title, Axis, grid, Text at random Place

```python
x = np.arange(0, 5, 0.1) # starting element, ending element and interval
y = x ** 3
y2 = x ** 2
# plt.plot(x,y, color ="black", marker = "^")
plt.plot(x,y, color ="black", linewidth = 5, label = "x^3") # with width of the desired line
plt.plot(x,y2, color ="red", linewidth = 4, label = "x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.axis([0,5, 0, 135])# it will take four arguments, first two for x axis and last two for y axis
plt.grid()
plt.text(2,40, "text", fontsize = 12) # it will take four arguments, the coordinates, what text do you want and size of the text
plt.legend()
plt.title("Mathematical function")
plt.show()
```
