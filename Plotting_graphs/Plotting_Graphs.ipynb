```python
import matplotlib.pyplot as plt

x =  [1,2,3]
y =  [2,4,6]
plt.scatter(x,y)

plt.show()
```

```python
plt.scatter(x,y)
plt.plot(x,y, "rp") # the first char of the string will decide the color and the second argument will decide type of the line
plt.show()
```

```python
plt.plot(x,y, "r-.")
plt.show()
```

```python
#plotting x^3
import numpy as np
x = np.array([1,2,3,4])
y = x**3

plt.plot(x,y, "r-.")
plt.show()
```

```python
x = np.arange(0, 5, 0.1) # starting element, ending element and interval
y = x ** 3
plt.plot(x,y, "b")
plt.show()
```
