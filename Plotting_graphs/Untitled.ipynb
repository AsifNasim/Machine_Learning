# Pie Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels =["A","B", "C", "D"]
sizes =[3,4,6,2]
colors = ["blue","purple","red", "pink"]
explode =[0.1, 0.2, 0, 0] #it will make a part of the pie distinct from all of them and it will take four float arguments
plt.title("Pie Graph Representation")
plt.pie(sizes, colors=colors, explode = explode, labels = labels, autopct="%.2f%%",counterclock=False, startangle =90) 
#by default the pie anticlockwise, autopct is autopercentage
plt.axis("equal") # it will make the pie purely circular
plt.show()
```
